```python
# In Colab, upload the repo or pip-install it from a GitHub URL after publication.
!pip install git+https://github.com/KNOWDYN/terafab-decision-twin.git
from terafab_decision_twin.schema import load_scenario, validate_scenario
from terafab_decision_twin.engine import run_scenario
from terafab_decision_twin.report import markdown_report
```

```python
import os
# Ensure the directory exists
os.makedirs('scenarios', exist_ok=True)

# Download the baseline scenario from the repository if it doesn't exist
if not os.path.exists('scenarios/baseline_2026.json'):
    !curl -L https://raw.githubusercontent.com/KNOWDYN/terafab-decision-twin/main/scenarios/baseline_2026.json -o scenarios/baseline_2026.json

scenario_path = 'scenarios/baseline_2026.json'
scenario = load_scenario(scenario_path)
errors = validate_scenario(scenario)
errors
```

```python
result = run_scenario(scenario)
result['summary']
```

```python
for gate in result['gates']:
    print(('PASS' if gate['passed'] else 'FAIL'), gate['name'], gate['margin'])
```

```python
report = markdown_report(result)
print(report[:3000])
```
